```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('accident_data.csv')
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
# Distribution of accidents based on weather conditions
plt.figure(figsize=(10, 6))
sns.countplot(x='Weather Conditions', data=df)
plt.title('Accidents based on Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('No of Vehichles')
plt.xticks(rotation=45)
plt.show()
```

```python

# Distribution of contributing factors
plt.figure(figsize=(10, 6))
sns.countplot(y='Contributing Factors', data=df)
plt.title('Distribution of Contributing Factors to Accidents')
plt.xlabel('No of Vehichles')
plt.ylabel('Contributing Factors')
plt.show()
```

```python
# Distribution of accidents based on road conditions
plt.figure(figsize=(10, 6))
sns.countplot(x='Road Conditions', data=df)
plt.title('Distribution of Accidents based on Road Conditions')
plt.xlabel('Road Conditions')
plt.ylabel('No of Vehichles')
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(12, 6))
sns.histplot(df['No of Vehichles'], bins=20, kde=True, color='skyblue')
plt.title('Types of collision')
plt.xlabel('No of Vehichles')
plt.ylabel('Vehicle Speed')
plt.xticks(rotation=45, ha='right')
plt.show()
```
